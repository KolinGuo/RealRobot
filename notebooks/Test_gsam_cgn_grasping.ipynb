```python
import real_robot.envs
import gym

image_obs_mode = "front"
camera_name = f"{image_obs_mode}_camera"
env = gym.make("GraspAnything-v0", image_obs_mode=image_obs_mode)
obs = env.reset()
env.agent.control_mode, env.agent.motion_mode
```

```python
import numpy as np
from PIL import Image

from contact_graspnet.estimator import CGNGraspEstimator
from grounded_sam_track.detection import Owl_ViT
from grounded_sam_track.segmentation import SAM
from grounded_sam_track.mask_tracking import XMem
from grounded_sam_track.pose_tracking import TAPIRPoseTracker
from grounded_sam_track.pose_tracking import BundleSDF

from grounded_sam_track.utils.visualization import draw_mask

from real_robot.utils.logger import get_logger
get_logger(log_file="/rl_benchmark/real_robot/notebooks/test.log")
_logger = get_logger("real_robot")

USE_BUNDLE_SDF = True

if __name__ == "__main__":
    grasp_estimator = CGNGraspEstimator("hor_sigma_0025", forward_passes=1)

    detector = Owl_ViT()
    segmentor = SAM()
    mask_tracker = XMem()
    if USE_BUNDLE_SDF:
        pose_tracker = BundleSDF()
    else:
        pose_tracker = TAPIRPoseTracker()

    rgb_frame = obs["image"][camera_name]["rgb"]
    depth_frame = obs["image"][camera_name]["depth"][..., 0]
    K = obs["camera_param"][camera_name]["intrinsic_cv"]
    T_world_cam = obs["camera_param"][camera_name]["cam2world_cv"]
    
    boxes, pred_indices, pred_scores = detector(rgb_frame, "mug")
    mask, pred_iou = segmentor(rgb_frame, boxes[pred_scores.argmax()])

Image.fromarray(draw_mask(rgb_frame, mask.cpu().numpy()))
```

```python
if __name__ == "__main__":
    from collections import defaultdict
    from copy import deepcopy
    import numpy as np
    import open3d as o3d
    from real_robot.utils.camera import depth2xyz, transform_points

    mask = mask_tracker(rgb_frame, mask, return_on_cpu=True).astype(bool)

    # # Initialize object pose using PCA
    # obj_xyz = depth2xyz(depth_frame, camera_params["intrinsic_cv"], 1.0)[mask]
    # obj_xyz = obj_xyz[obj_xyz[:, 2] > 0]
    # obj_center = obj_xyz.mean(axis=0)
    # U, S, Vh = np.linalg.svd(obj_xyz - obj_center)
    # if np.linalg.det(Vh) < 0:
    #     Vh[2, :] *= -1
    # T_cam_obj = np.eye(4)
    # T_cam_obj[:3, :3] = Vh.T
    # T_cam_obj[:3, 3] = obj_center
    # T_obj_cam = np.linalg.inv(T_cam_obj)
    # Initialize object pose with position only
    obj_xyz = depth2xyz(depth_frame, K, 1.0)[mask]
    obj_xyz = obj_xyz[obj_xyz[:, 2] > 0]
    obj_center = obj_xyz.mean(axis=0)
    T_world_obj = np.eye(4)
    T_world_obj[:3, 3] = transform_points(obj_center, T_world_cam)
    T_obj_cam = np.linalg.inv(T_world_obj) @ T_world_cam
    pose_tracker.initialize_T_obj_cam(T_obj_cam)
    prev_T_obj_cam = T_obj_cam

    # T_obj_cam = pose_tracker(rgb_frame, depth_frame, camera_params["intrinsic_cv"], mask)

    pred_grasps_cam, scores, contact_pts, q_vals = grasp_estimator.predict_grasps(
        depth=depth_frame, seg=mask,
        K=K, local_regions=False, filter_grasps=True
    )
    pred_grasps_world = T_world_cam @ pred_grasps_cam

    grasp_estimator.visualize_grasps(env.visualizer.o3d_vis, pred_grasps_world, scores, q_vals)
    T_gripper_tcp = np.eye(4)
    T_gripper_tcp[2, -1] = 0.172

    # Visualization
    obs_dict = defaultdict(list)

    xyz_image = depth2xyz(depth_frame, K, 1.0)
    xyz_image = transform_points(xyz_image, T_world_cam)

    obs_dict["camera_names"].append(camera_name)
    obs_dict["color_images"].append(rgb_frame)
    obs_dict["depth_images"].append(depth_frame)
    obs_dict["mask"].append(mask)
    obs_dict["xyz_images"].append(xyz_image)
    obs_dict["obj_frame_mesh"].append(o3d.geometry.TriangleMesh.create_coordinate_frame(size=0.5).transform(
        T_world_cam @ np.linalg.inv(T_obj_cam)
    ))

    env.visualizer.show_observation(**obs_dict)
    env.visualizer.o3d_vis.toggle_pause(True)
    env.visualizer.render()
```
